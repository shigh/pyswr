```python
import numpy as np
from region import *
from recursive import *
from utils import *
import scipy.sparse as sparse
import scipy.sparse.linalg as splinalg
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
nt = 100
nx = 100
p=1.

x_max = 4.*np.pi
dt = 1./(nt-1)
dx = x_max/(nx-1)

x0_full  = np.linspace(0, x_max, nx)
f0_full  = np.sin(x0_full)

expected = np.zeros((nt, nx))
exact_solver = ImplicitSolver1DRec(dt, nx, dx, False, False, p=p)
exact_solver.x[:] = f0_full
for t in range(1, nt):
    exact_solver.solve()
    expected[t] = exact_solver.x.copy()

n_reg  = 2
f_vals = region_views(f0_full, n_reg, 0)
x_vals = region_views(x0_full, n_reg, 0)

# Number of iterations
N = 16
tol = .002
```

## Schwartz Waveform

```python
# Keep track of errors as a function of iterations and time
wave_errors = np.zeros((N, nt))

# Build solvers
solvers = [ImplicitSolver1DRec(dt, len(f_vals[i]), dx, i>0, i<n_reg-1, p=p)
           for i in range(n_reg)]

# Build regions
regions = [RecRegion(nt, len(s.x), p=p) for s in solvers]
for r in regions:
    for t in range(0, nt):
        r.add_slice(t)
for r, f in zip(regions, f_vals):
        r.slices[0][:] = f
    
# Waveform iterations
for itr in range(N):
    
    # Reset solvers for next iteration
    for s, r in zip(solvers, regions):
        s.x[:] = r.slices[0]
    
    # Apply solvers over each time step
    for t in range(1, nt):
        for s, r in zip(solvers, regions):
            s.g = [r.g[0][0][t], r.g[0][-1][t]]
            s.solve()
            r.update_cols(t, s.x)
            
    # Calc error at each step     
    for t in range(1, nt):
        comb  = combine_arrays([r.slices[t] for r in regions], nx)
        error = np.max(np.abs(expected[t]-comb))
        wave_errors[itr, t] = error
        
    # Update g values
    for j in range(n_reg):
        s = solvers[j]
        r = regions[j]
        if s.has_right:
            r.g[0][-1] = regions[j+1].send_g(0, 0)
        if s.has_left:
            r.g[0][0]  = regions[j-1].send_g(0, -1)
```

## Schwartz Iteration

```python
# Keep track of errors as a function of iterations and time
schwartz_errors = np.zeros((N, nt))

# Build solvers
solvers = [ImplicitSolver1DRec(dt, len(f_vals[i]), dx, 
                               i>0, i<n_reg-1, p=p)
           for i in range(n_reg)]

# Build regions
regions = [RecRegion(nt, len(s.x), p=p) for s in solvers]
for r in regions:
    for t in range(0, nt):
        r.add_slice(t)
for r, f in zip(regions, f_vals):
        r.slices[0][:] = f
    
# Waveform iterations
for t in range(1, nt):

    # Apply solvers over each time step
    for itr in range(N):
        
        # Reset solver for next iteration
        for s, r in zip(solvers, regions):
            s.x[:] = r.slices[t-1]
        
        # Solve in each region
        for s, r in zip(solvers, regions):
            s.g = [r.g[0][0][t], r.g[0][-1][t]]
            s.solve()
            r.update_cols(t, s.x)
            
        # Calc error after each
        comb  = combine_arrays([r.slices[t] for r in regions], nx)
        error = np.max(np.abs(expected[t]-comb))
        schwartz_errors[itr, t] = error
        
        # Update g values
        for j in range(n_reg):
            s = solvers[j]
            r = regions[j]
            if s.has_right:
                r.g[0][-1] = regions[j+1].send_g(0, 0)
            if s.has_left:
                r.g[0][0]  = regions[j-1].send_g(0, -1)
```

## Schwartz Waveform with resets

```python
def waveform(n_resets):
    """Wrap waveform iteration.
    
    Pulls general settings from global vars
    """
    # Build solvers
    solvers = [ImplicitSolver1DRec(dt, len(f_vals[i]), 
                                   dx, i>0, i<n_reg-1, p=p)
               for i in range(n_reg)]
    
    # Build regions
    regions = [RecRegion(nt, len(s.x), p=p) for s in solvers]
    for r in regions:
        for t in range(0, nt):
            r.add_slice(t)
    for r, f in zip(regions, f_vals):
            r.slices[0][:] = f
    
    errors = np.zeros((N, nt))
    t_ranges = []
    steps = int(float(nt)/n_resets)
    for i in range(n_resets):
        t_ranges.append((i*steps, (i+1)*steps))
    t_ranges[-1] = (t_ranges[-1][0], nt)
    t_ranges[0] = (1, t_ranges[0][1])
    
    for (t0, t1) in t_ranges:
        # Waveform iterations
        for itr in range(N):
            
            # Reset solvers for next iteration
            for s, r in zip(solvers, regions):
                s.x[:] = r.slices[t0-1]
            
            # Apply solvers over each time step
            for t in range(t0, t1):
                for s, r in zip(solvers, regions):
                    s.g = [r.g[0][0][t], r.g[0][-1][t]]
                    s.solve()
                    r.update_cols(t, s.x)
                    
            # Calc error at each step     
            for t in range(t0, t1):
                comb  = combine_arrays([r.slices[t] 
                                        for r in regions], nx)
                error = np.max(np.abs(expected[t]-comb))
                errors[itr, t] = error
                
            # Update g values
            for j in range(n_reg):
                s = solvers[j]
                r = regions[j]
                if s.has_right:
                    r.g[0][-1] = regions[j+1].send_g(0, 0)
                if s.has_left:
                    r.g[0][0]  = regions[j-1].send_g(0, -1)
    return errors
                    
reset_errors = waveform(2)
print np.max(np.abs(waveform(1)-wave_errors))
print np.max(np.abs(waveform(nt)-schwartz_errors))
```

```python
reset_errors = waveform(10)
plt.plot(np.log(reset_errors[-1, 1:]))
plt.plot(np.log(schwartz_errors[-1, 1:]))
```

```python
err_diffs = []
for i in range(1, 10+1):
    reset_err = waveform(i)[-1, 1:]
    err_diff = np.max(np.abs(schwartz_errors[-1, 1:]-reset_err))
    err_diffs.append(err_diff)
plt.plot(np.log(err_diffs))  
```

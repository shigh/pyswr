## Single Step Tests 1D

```python
nt = 100
nx = 100

dt = 1./(nt-1)
dx = 2.*np.pi/(nx-1)

x0_full = np.linspace(0, 2*np.pi, nx)
f0_full = np.sin(x0_full)

f_left  = f0_full[:50]
f_right = f0_full[49:]
x_left  = x0_full[:50] 
x_right = x0_full[49:]
```

```python
# Number of iterations
N = 7
sl = ImplicitSolver1DRec(dt, len(f_left), dx, False, True)
sr = ImplicitSolver1DRec(dt, len(f_right), dx, True, False)
sl.x[:] = f_left
sr.x[:] = f_right

for i in range(N):
    
    # Apply solver
    for s in [sl, sr]:    
        s.solve()
        
    # Update g values
    sl.g[1] = sr.send_g()[0]
    sr.g[0] = sl.send_g()[1]
        
    # Reset solver for next iteration
    if i<N-1:
        sl.x[:] = f_left
        sr.x[:] = f_right
```

```python
plot(x_left, sl.x)
plot(x_right, sr.x)
```

```python
nt = 2
nx = 100
x_max = 4.*np.pi
dt = 1./(nt-1)
dx = x_max/(nx-1)

x0_full  = np.linspace(0, x_max, nx)
f0_full  = np.sin(x0_full)
expected = np.exp(-1.)*f0_full
n_reg  = 5
f_vals = region_views(f0_full, n_reg, 0) 
x_vals = region_views(x0_full, n_reg, 0)
```

```python
# Number of iterations
N = 4

solvers = [ImplicitSolver1DRec(dt, len(f_vals[i]), dx, i>0, i<n_reg-1)
           for i in range(n_reg)]

for s, f in zip(solvers, f_vals):
    s.x[:] = f

for i in range(N):
    
    # Apply solvers
    for s in solvers:    
        s.solve()
        
    # Update g values
    updated_g = [s.send_g() for s in solvers]
    for j in range(n_reg):
        s = solvers[j]
        if s.has_right:
            s.g[1] = updated_g[j+1][0]
        if s.has_left:
            s.g[0] = updated_g[j-1][1]
        
    # Reset solver for next iteration
    if i<N-1:
        for s, f in zip(solvers, f_vals):
            s.x[:] = f
        
```

```python
plot(x0_full, expected)
for x, s in zip(x_vals, solvers):
    plot(x, s.x)
```

## Full Waveform 1D

```python
nt = 100
nx = 100

x_max = 4.*np.pi
dt = 1./(nt-1)
dx = x_max/(nx-1)

x0_full  = np.linspace(0, x_max, nx)
f0_full  = np.sin(x0_full)
expected = np.exp(-1.)*f0_full
n_reg  = 2
f_vals = region_views(f0_full, n_reg, 0)
x_vals = region_views(x0_full, n_reg, 0)
```

```python
# Number of iterations
N = 20

# Build solvers
solvers = [ImplicitSolver1DRec(dt, len(f_vals[i]), dx, i>0, i<n_reg-1)
           for i in range(n_reg)]

# Build regions
regions = [RecRegion(nt, len(s.x)) for s in solvers]
for r, f in zip(regions, f_vals):
        r.slices[0][:] = f
        
for i in range(N):
    
    # Reset solvers for next iteration
    for s, r in zip(solvers, regions):
        s.x[:] = r.slices[0]
    
    # Apply solvers over each time step
    for t in range(1, nt):
        for s, r in zip(solvers, regions):
            s.g = [r.g[0][0][t], r.g[0][-1][t]]
            s.solve()
            r.update_cols(t, s.x)
        
    # Update g values
    for j in range(n_reg):
        s = solvers[j]
        r = regions[j]
        if s.has_right:
            r.g[0][-1] = regions[j+1].send_g(0, 0)
        if s.has_left:
            r.g[0][0]  = regions[j-1].send_g(0, -1)
        
        
        
```

```python
plot(x0_full, expected)
for x, r in zip(x_vals, regions):
    plot(x, r.slices[nt-1])
```

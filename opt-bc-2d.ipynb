## Single Step Tests 2D

```python
nt = 2
nx = 100
ny = 100

x_max = 3.*np.pi
y_max = 3.*np.pi
dt = 1./(nt-1)
dx = x_max/(nx-1)
dy = y_max/(ny-1)

x0_full, y0_full = np.meshgrid(np.linspace(0, x_max, nx), np.linspace(0, y_max, ny))
f0_full  = np.sin(x0_full)*np.sin(y0_full)
expected = np.exp(-2.)*f0_full
```

```python
x_left   = x0_full[:50, :50]
x_right  = x0_full[:50, 49:]
y_left   = y0_full[:50, :50]
y_right  = y0_full[:50, 49:]
f_left   = f0_full[:50, :50]
f_right  = f0_full[:50, 49:]

x_leftu  = x0_full[49:, :50]
x_rightu = x0_full[49:, 49:]
y_leftu  = y0_full[49:, :50]
y_rightu = y0_full[49:, 49:]
f_leftu  = f0_full[49:, :50]
f_rightu = f0_full[49:, 49:]
```

```python
# Number of iterations
N = 40

sl  = ImplicitSolver2DRec(dt, f_left.shape[0],  dy, f_left.shape[1],  dx, 
                          False, True, True, False)
sr  = ImplicitSolver2DRec(dt, f_right.shape[0], dy, f_right.shape[1], dx, 
                          True, False, True, False)
slu = ImplicitSolver2DRec(dt, f_leftu.shape[0],  dy, f_leftu.shape[1],  dx, 
                          False, True, False, True)
sru = ImplicitSolver2DRec(dt, f_rightu.shape[0], dy, f_rightu.shape[1], dx, 
                          True, False, False, True)


sl.x[:] = f_left
sr.x[:] = f_right
slu.x[:] = f_leftu
sru.x[:] = f_rightu

for i in range(N):
    
    # Apply solver
    for s in [sl, sr, slu, sru]:    
        s.solve() 
        
    # Update g values
    sl.g[1][-1] = sr.send_g()[1][0]
    sl.g[0][-1] = slu.send_g()[0][0]
    
    sr.g[1][0] = sl.send_g()[1][-1] 
    sr.g[0][-1] = sru.send_g()[0][0]
    
    slu.g[1][-1] = sru.send_g()[1][0]
    slu.g[0][0]  = sl.send_g()[0][-1]
    
    sru.g[1][0] = slu.send_g()[1][-1]
    sru.g[0][0] = sr.send_g()[0][-1]
    
    # Reset solver for next iteration
    if i<N-1:
        sl.x[:]  = f_left
        sr.x[:]  = f_right
        slu.x[:] = f_leftu
        sru.x[:] = f_rightu
        
```

```python
fig = plt.figure()
fig.set_size_inches(14, 10)
ax = fig.add_subplot(111, projection='3d')
cs, rs = 1, 1
ax.plot_wireframe(x_left, y_left, sl.x, cstride=cs, rstride=rs)
ax.plot_wireframe(x_right, y_right, sr.x, color='g', cstride=cs, rstride=rs)
ax.plot_wireframe(x_leftu, y_leftu, slu.x, color='c',cstride=cs, rstride=rs)
ax.plot_wireframe(x_rightu, y_rightu, sru.x, color='m', cstride=cs, rstride=rs)
ax.plot_wireframe(x0_full, y0_full, expected, color='r', cstride=cs, rstride=rs)
plt.show()
```

## Full Waveform 2D

```python
nt = 100
nx = 200
ny = 100

x_max = 3.*np.pi
y_max = 3.*np.pi
dt = 1./(nt-1)
dx = x_max/(nx-1)
dy = y_max/(ny-1)

x0_full, y0_full = np.meshgrid(np.linspace(0, x_max, nx), np.linspace(0, y_max, ny))
f0_full  = np.sin(x0_full)*np.sin(y0_full)
expected = np.exp(-2.)*f0_full

n_reg_x = 4
n_reg_y = 3

f_vals = region_views_2d(f0_full, n_reg_y, n_reg_x)
x_vals = region_views_2d(x0_full, n_reg_y, n_reg_x)
y_vals = region_views_2d(y0_full, n_reg_y, n_reg_x)
```

```python
# Number of iterations
N = 20

# Build solvers
solvers = [[ImplicitSolver2DRec(dt, f_vals[i][j].shape[0], dy, 
                                    f_vals[i][j].shape[1], dx, 
                                    j>0, j<n_reg_x-1,
                                    i<n_reg_y-1, i>0)
            for j in range(n_reg_x)]
            for i in range(n_reg_y)]
```

```python
# Build regions
regions = [[RecRegion(nt, s.x.shape) for s in row_s]
            for row_s in solvers]

for row_r, row_f in zip(regions, f_vals):
    for r, f in zip(row_r, row_f):
        r.slices[0][:] = f
```

```python
for i in range(N):
    
    # Reset solvers for next iteration
    for row_s, row_r in zip(solvers, regions):
        for s, r in zip(row_s, row_r):
            s.x[:] = r.slices[0]

    # Apply solvers over each time step
    for t in range(1, nt):
        for row_s, row_r in zip(solvers, regions):
            for s, r in zip(row_s, row_r):
                s.g = [[r.g[0][0][t], r.g[0][-1][t]],
                       [r.g[1][0][t], r.g[1][-1][t]]]
                s.solve()
                r.update_cols(t, s.x)
        
    # Update g values
    for i in range(n_reg_y):
        for j in range(n_reg_x):
            s = solvers[i][j]
            r = regions[i][j]
            if s.has_right:
                r.g[1][-1] = regions[i][j+1].send_g(1, 0)
            if s.has_left:
                r.g[1][0]  = regions[i][j-1].send_g(1,-1)
            if s.has_north:
                r.g[0][-1] = regions[i+1][j].send_g(0, 0)
            if s.has_south:
                r.g[0][0]  = regions[i-1][j].send_g(0,-1)
```

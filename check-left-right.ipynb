## Full Waveform 2D

```python
nt = 10
nx = 100
ny = 100

x_max = 3.*np.pi
y_max = 3.*np.pi
dt = 1./(nt-1)
dx = x_max/(nx-1)
dy = y_max/(ny-1)

x0_full, y0_full = np.meshgrid(np.linspace(0, x_max, nx), 
                               np.linspace(0, y_max, ny))
f0_full  = np.sin(x0_full)*np.sin(y0_full)

expected_solver = ImplicitSolver2DRec(dt, ny, dy, nx, dx,
                                      False, False, False, False)
expected_solver.x[:] = f0_full
for i in range(1, nt):
    expected_solver.solve()
    
expected = expected_solver.x

n_reg_x = 2
n_reg_y = 2

f_vals = region_views_2d(f0_full, n_reg_y, n_reg_x)
x_vals = region_views_2d(x0_full, n_reg_y, n_reg_x)
y_vals = region_views_2d(y0_full, n_reg_y, n_reg_x)
```

```python
# Number of iterations
N = 20

# Build solvers
solvers = [[ImplicitSolver2DRec(dt, f_vals[i][j].shape[0], dy, 
                                    f_vals[i][j].shape[1], dx, 
                                    j>0, j<n_reg_x-1,
                                    i<n_reg_y-1, i>0)
            for j in range(n_reg_x)]
            for i in range(n_reg_y)]
```

```python
# Build regions
regions = [[RecRegion(nt, s.x.shape) for s in row_s]
            for row_s in solvers]

for row_r, row_f in zip(regions, f_vals):
    for r, f in zip(row_r, row_f):
        r.slices[0][:] = f
```

```python
for i in range(N):
    
    # Reset solvers for next iteration
    for row_s, row_r in zip(solvers, regions):
        for s, r in zip(row_s, row_r):
            s.x[:] = r.slices[0]

    # Apply solvers over each time step
    for t in range(1, nt):
        for row_s, row_r in zip(solvers, regions):
            for s, r in zip(row_s, row_r):
                s.g = [[r.g[0][0][t], r.g[0][-1][t]],
                       [r.g[1][0][t], r.g[1][-1][t]]]
                s.solve()
                r.update_cols(t, s.x)
        
    # Update g values
    for i in range(n_reg_y):
        for j in range(n_reg_x):
            s = solvers[i][j]
            r = regions[i][j]
            if s.has_right:
                r.g[1][-1] = regions[i][j+1].send_g(1, 0)
            if s.has_left:
                r.g[1][0]  = regions[i][j-1].send_g(1,-1)
            if s.has_north:
                r.g[0][-1] = regions[i+1][j].send_g(0, 0)
            if s.has_south:
                r.g[0][0]  = regions[i-1][j].send_g(0,-1)
```

```python
fig = plt.figure()
fig.set_size_inches(14, 10)
ax = fig.add_subplot(111, projection='3d')
cs, rs = 1, 1
for i in range(n_reg_y):
    for j in range(n_reg_x):
        ax.plot_wireframe(x_vals[i][j], y_vals[i][j], 
                          regions[i][j].slices[nt-1], 
                          cstride=cs, rstride=rs, color='r')
ax.plot_wireframe(x0_full, y0_full, expected, color='b', 
                  cstride=cs, rstride=rs)
#ax.plot_wireframe(x0_full, y0_full, x, color='r', cstride=cs, rstride=rs)
plt.show()
```
